```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#plt.style.use('default')
```

```python
# Plotting a simple function (line plot)
price= [48000, 54000, 57000, 49000, 47000,45000]
year= [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
batsman= pd.read_csv("E://Technical//DS_AI_ML//campusx//datasets//sharma-kohli.csv")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"])
```

```python
# Plotting multiple plots
plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])
```

```python
# xlabel and ylabel title
plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# color change in matplotlib plots
plt.plot(batsman["index"], batsman["V Kohli"], color= "green")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# linestyle -> dashed
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashed")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashed")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# linestyle -> dotted
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dotted")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dotted")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# linestyle -> dashdot
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# linewidth
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3)
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2)
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# marker
# marker= "D", "+", ".", "<", ">", "o"
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3, marker= "D")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2, marker= "o")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# markersize
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3, marker= "D", markersize= 10)
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2, marker= "o", markersize= 5)
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
```

```python
# labels and plt.legend(), we can change the location of the label so we can add that as per our requirement by passing 
# arguments inside plt.legend()
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3, marker= "D", markersize= 10, label= "Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2, marker= "o", markersize= 5, label= "Rohit")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
plt.legend()
```

```python
# Plotting a simple function (line plot) with xlim and ylim
price= [48000, 54000, 57000, 49000, 47000,45000]
year= [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.xlim(2017, 2019)
plt.ylim(0, 75000)
```

```python
# plt.grid()
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3, marker= "D", markersize= 10, label= "Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2, marker= "o", markersize= 5, label= "Rohit")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
plt.legend()
plt.grid()
```

```python
# plt.show() is used to plot the graphs omn multiple ide like py charm, vs code etc.
plt.plot(batsman["index"], batsman["V Kohli"], color= "green", linestyle= "dashdot", linewidth= 3, marker= "D", markersize= 10, label= "Virat")
plt.plot(batsman["index"], batsman["RG Sharma"], color= "black", linestyle= "dashdot", linewidth= 2, marker= "o", markersize= 5, label= "Rohit")
plt.title("Rohit Sharma v/s Virat Kohli IPL Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs scored")
plt.legend()
plt.grid()
plt.show()
```

## Scatter plot
* Bivariate Analysis
* numerical vs numerical
* Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)

plt.scatter(x, y)
```

```python
# plt.scatter on pandas data
df= pd.read_csv("E://Technical//DS_AI_ML//campusx//datasets//batter.csv")
df= df.head(50)
df
```

```python
# size
# slower scatterplot representation
tips= sns.load_dataset("tips")
plt.scatter(tips["total_bill"], tips["tip"], s= tips["size"]*20)
```

## Bar Chart

* Bivariate Analysis
* Numerical vs Categorical
* Use case - Aggregate analysis of groups

```python
# simple bar chart
children= [10, 20 ,40, 10, 30]
colors= ["red", "blue", "green", "yellow", "pink"]
plt.bar(colors, children, color= "black")
```

```python
# horizontal bar chart
plt.barh(colors, children, color= "black")
```

```python
df= pd.read_csv("E://Technical//DS_AI_ML//campusx//datasets//batsman_season_record.csv")
df
```

```python
np.arange(df.shape[0])
```

```python
# plt.xticks()
# using np.arange(df.shape()) on x axis
plt.bar(np.arange(df.shape[0])- 0.2, df["2015"], width= 0.2, color= "green")
plt.bar(np.arange(df.shape[0]), df["2016"], width= 0.2, color= "red")
plt.bar(np.arange(df.shape[0])+ 0.2, df["2017"], width= 0.2, color= "blue")
# plt.xticks() is used to replace the np.arange(df.shape[0]) value with df["batsman"] 
plt.xticks(np.arange(df.shape[0]), df["batsman"])
# plt.show() is used to remove the unwanted message on the output console
plt.show()
```

```python
# plt.xticks(rotation= "vertical") for more readibility in x axis, if the value of x axis is large
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']
plt.bar(colors, children, color= "black")
plt.xticks(rotation= "vertical")
```

```python
# stacked bar chart
plt.bar(df["batsman"], df["2017"], label="2017")
plt.bar(df["batsman"], df["2016"], bottom= df["2017"], label="2016")
plt.bar(df["batsman"], df["2015"], bottom= df["2017"] + df["2016"], label="2015")

plt.legend()
plt.show()
```

```python
# linear scale (1-2 is equal to 2-3 and 100-200 is equal to 200-300)
# there are almost 10000 numbers in 30-40 bins and because of that it is not able to show us the numbers for the 
# remaining bins.
plt.hist(arr, bins= [10, 20, 30, 40, 50, 60, 70])
plt.show()
```

```python
# logarithmic scale (1-100 is equal to 100-1000)
# in order to avoid the problem which we have faced in the above cell we will use logarithmic scale insted of linear scale.
plt.hist(arr, bins= [10, 20, 30, 40, 50, 60, 70], log= True)
plt.show()
```

## Pie chart
* Univariate/Bivariate Analysis
* Categorical vs numerical
* Use case - To find contibution on a standard scale

```python
# sample data
marks= [23, 45, 100, 20, 49]
subject= ["eng", "science", "maths", "sst", "hindi"]
plt.pie(marks, labels= subject)
plt.show()
```

```python
# dataset
df= pd.read_csv("E://Technical//DS_AI_ML//campusx//datasets//gayle-175.csv")
df
```

```python
# autopct() for percentage
plt.pie(df["batsman_runs"], labels= df["batsman"], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df["batsman_runs"], labels= df["batsman"], autopct='%0.1f%%', colors= ["cyan", "red", "brown", "blue", "pink", "yellow"])
plt.show()
```

```python
# explode 
plt.pie(df["batsman_runs"], labels= df["batsman"], autopct='%0.1f%%', explode= [0.3, 0, 0, 0, 0, 0.1])
plt.show()
```

```python
# explode with shadow
plt.pie(df["batsman_runs"], labels= df["batsman"], autopct='%0.1f%%', explode= [0.3, 0, 0, 0, 0, 0.1], shadow= True)
plt.show()
```
